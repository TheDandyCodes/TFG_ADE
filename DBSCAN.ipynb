```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msn
import seaborn as sns
import plotly.express as px
```

```python
from sklearn.datasets import make_blobs
X_blobs, Y_blobs= make_blobs(n_samples=500, random_state=8, n_features=2, cluster_std=2.5)
plt.scatter(X_blobs[:,0], X_blobs[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
sil = []
min_samples_values = [2, 3, 4, 5, 6, 7, 8, 9, 10]
for min_samples in min_samples_values:
    dbscan = DBSCAN(eps=3, min_samples=min_samples)
    dbscan.fit(X_blobs)
    clusters = dbscan.labels_
    silhouette_score_value = silhouette_score(X_blobs, clusters)
    sil.append(silhouette_score_value)
plt.plot(min_samples_values, sil)
plt.show()
```

```python
dbscan = DBSCAN(eps=3, min_samples=4)
dbscan.fit(X_blobs)
plt.scatter(X_blobs[:,0], X_blobs[:,1],c=dbscan.labels_)
plt.show()
```
